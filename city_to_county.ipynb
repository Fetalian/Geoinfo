Read Seru data and correct encoding. 

```python
IAR= pd.read_stata(dta_path)
# replace non-latin characters with '', which arise due to encoding problem(unicode source while pd.read_stata use latin-1)
IAR.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
```

```python
def find_county(s):
    try: 
        res=search.by_city_and_state(s.city, s.state)
        zipcode=res[0]
        return zipcode.county
    except:
        pass
#         print('s%, s% not matched'.format(s.city,s.state))
        return np.nan
```

```python
test=IAR.iloc[0:20,0:5]

test.loc[:,'county']=test.swifter.apply(find_county,axis=1)
```

```python
IAR.loc[:,'county']=IAR.iloc[:,0:5].swiapply(find_county,axis=1)
```
